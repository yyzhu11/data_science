## Leetcode 1137. N-th Tribonacci Number

The Tribonacci sequence Tn is defined as follows:

T0 = 0, T1 = 1, T2 = 1, and Tn+3 = Tn + Tn+1 + Tn+2 for n >= 0.Given n, return the value of Tn.

- Example 1:
Input: n = 4
Output: 4

Explanation:
T_3 = 0 + 1 + 1 = 2
T_4 = 1 + 1 + 2 = 4

- Example 2:
Input: n = 25
Output: 1389537

````python
def Tribonacci(n):
    if n==0:
        return 0
    if n==1:
        return 1
    if n==2:
        return 1
    if n >=3:
        return Tribonacci(n-1) + Tribonacci(n-2) + Tribonacci(n-3)
````

Description:
Given two binary strings, return their sum (also a binary string).

The input strings are both non-empty and contains only characters 1 or 0.

- Example 1:

Input: a = "11", b = "1"
Output: "100"

- Example 2:

Input: a = "1010", b = "1011"
Output: "10101"

````python
def addBinary(a, b):
    if not a and not b:
        return ''
    if not a:
        return b
    if not b:
        return a
    if a[-1] =='0' and b[-1] =='0':
        return addBinary(a[:-1], b[:-1]) + '0'
    if (a[-1] =='1' and b[-1] =='0') or (a[-1] =='0' and b[-1] =='1'):
        return addBinary(a[:-1], b[:-1]) + '1'
    if a[-1] =='1' and b[-1] =='1':
        return addBinary(addBinary(a[:-1], b[:-1]), '1') + '0'
````

# Binary Tree Functions

````python
#class node
from collections import deque
class TreeNode():
    def __init__(self, val=0, left=None, right=None):
        self.left = left
        self.right = right
        self.val = val

def to_binary_tree(items):
    if not items:
        return None
    it = iter(items)
    root = TreeNode(next(it))
    q = [root]
    for node in q:
        val = next(it, None)
        if val is not None:
            node.left = TreeNode(val)
            q.append(node.left)        
        val = next(it, None)
        if val is not None:
            node.right = TreeNode(val)
            q.append(node.right)
    return root

#Level Order Traversal
def print_binary_tree(root):
    queue = deque()
    result = []
    queue.append(root)
    while not all(item is None for item in queue):
        node = queue.popleft()
        if node:
            result.append(node.val)       
            if node.left is not None:
                #result.append(node.left.val)
                #print('add left:', node.left.val)
                queue.append(node.left)
            else:
                #print('add left: None')
                queue.append(None)
            if node.right is not None:
                #result.append(node.right.val)
                #print('add right:', node.right.val)
                queue.append(node.right)
            else:
                #print('add right: None')
                queue.append(None)
        else:
            result.append(None)
        
    return result     
````

- Example 1:

Output: true
- Example 2:



Input: root = [1,2,2,null,3,null,3]

````python
def isMirror(left, right):
    if left is None and right is None:
        return True
    if left and not right:
        return False
    if right and not left:
        return False
    if right.val != left.val:
        return False
    return isMirror(left.right, right.left) and isMirror(left.left, right.right)

def symmetricTree(root):
    if not root:
        return True
    return isMirror(root.left, root.right)
````

Description:
Given an array where elements are sorted in ascending order, convert it to a height balanced BST.

For this problem, a height-balanced binary tree is defined as a binary tree 
in which the depth of the two subtrees of every node never differ by more than 1.

- Example:
Given the sorted array: [-10,-3,0,5,9],One possible answer is: [0,-3,9,-10,null,5], 
which represents the following height balanced BST:

```
      0
    /   \
   -3    9
   /    / 
 -10    5
```
Input: nums = [-10,-3,0,5,9]
Output: [0,-3,9,-10,null,5]
Explanation: [0,-10,5,null,-3,null,9] is also accepted

```
        0
       /  \ 
     -10   5
        \   \
        -3   9


```



- Example

Input: nums = [1,3]
Output: [3,1]
```
     1             3
      \           /
       3         1 
```

Explanation: [1,null,3] and [3,1] are both height-balanced BSTs

````python
def convertBST(nums):
    if not nums:
        return None
    #if len(nums) == 1:
    #    return TreeNode(nums[0])
    left, right = 0, len(nums)-1
    mid = (left + right)//2
    root = TreeNode(nums[mid])
    root.left = convertBST(nums[:mid])
    root.right = convertBST(nums[mid+1:])
    return root    
````

```

          /     \
         3      3
        /        \

```
Input: [1,2,2,3,None,None,3,4,None,None,4] output = False

````python
def height(node):
    if node is None:
        return 0
    right_height = height(node.left)
    left_height = height(node.right) 
    if right_height ==-1 or left_height ==-1 or abs(right_height - left_height) >1:
        return -1
    return 1 + max(left_height, right_height)

def isBalacedBT(root):
    if root is None:
        return True
    return height(root) >= 0
````

````python
def sumLeftLeaves(root):
    if root is None:
        return 0
    sum_leave = sumLeftLeaves(root.right)
    if root.left and root.left.right is None and root.left.left is None:
        sum_leave += root.left.val
    else:
        sum_leave += sumLeftLeaves(root.left)
    return sum_leave
````

## Leetcode 617. Merge Two Binary Trees

Description:
Given two binary trees and imagine that when you put one of them to cover the other, some nodes of the two trees are overlapped while the others are not.

You need to merge them into a new binary tree. The merge rule is that if two nodes overlap, then sum node values up as the new value of the merged node. Otherwise,

the NOT null node will be used as the node of new tree.

Note:
The merging process must start from the root nodes of both trees.

Example 1:
Input: 

	Tree 1                     Tree 2                  
          1                         2                             
         / \                       / \                            
        3   2                     1   3                        
       /                           \   \                      
      5                             4   7        
      
Output: 
Merged tree:

	     3
	    / \
	   4   5
	  / \   \ 
	 5   4   7

````python
def mergeTree(root1, root2):
    if not root1 and not root2:
        return None
    if not root1 and root2:
        return root2
    if not root2 and root1:
        return root1
    if root1 and root2:
        node = TreeNode(root1.val + root2.val)
        node.left = mergeTree(root1.left, root2.left)
        node.right = mergeTree(root1.right, root2.right)
        return node
````

## Leetcode 700. Search in a Binary Search Tree

Description:
Given the root node of a binary search tree (BST) and a value. 
You need to find the node in the BST that the node’s value equals the given value.

Return the subtree rooted with that node. If such node doesn’t exist, you should return NULL.

- Example:
Given the tree:
```
    4
   / \
  2   7
 / \
1   3
```

- Input: root = [4,2,7,1,3], val = 2, Output: [2,1,3]

You should return this subtree:
```
     2     
    / \   
   1   3
```

- Input: root = [4,2,7,1,3], val = 5, Output: []

In the example above, if we want to search the value 5, 
since there is no node with value 5, we should return NULL.

Note that an empty tree is represented by NULL,
