# leetcode 136. Single Number

Given a non-empty array of integers, every element appears twice except for one. Find that single one.

Note:
Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

- Example: Input: [2,2,1], Output: 1


```python
# use math
def singleNumber(nlist):
    mySet = set(nlist)
    return sum(mySet)*2 - sum(nlist)
```

```python
# use dictionary
def singleNumber2(nlist):
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict.keys():
        if myDict[i] == 1:
            return i
```

```python
# use
def singleNumber3(nlist):
    if not nlist:
        return None
    result = nlist[0]
    for i in nlist[1:]:
        result = result ^ i
    return result
```

```python
singleNumber3([2,2,1])
```

# leetcode 169. Majority Element

Description:
Given an array of size n, find the majority element. The majority element is the element that appears more than ⌊ n/2 ⌋ times.

Note:
You may assume that the array is non-empty and the majority element always exist in the array.

- Example:
  Input: [3,2,3], Output: 3

```python
def majorityElement(nlist):
    if not nlist:
        return None
    n = len(nlist)
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict.keys():
        if myDict[i] > n//2:
            return i
```

```python
majorityElement([3,2,3])
```

# leetcode 217. Contains Duplicate

- Description:
Given an array of integers, find if the array contains any duplicates.

Note:
Your function should return true if any value appears at least twice in the array, and it should return false
if every element is distinct.

- Example: Input: [1,2,3,1], Output: true

```python
def containDuplicate(nlist):
  if not nlist:
    return False
  myDict = {}
  for i in nlist:
    if i in myDict:
      return True
    else:
      myDict[i] = myDict.get(i, 0)+1
  return False
```

# Leetcode 1. Two Sum

- Description:
Given an array of integers, return indices of the two numbers such that they add up to a specific target.

Note:
You may assume that each input would have exactly one solution, and you may not use the same element twice.

- Example:
Given nums = [2, 7, 11, 15], target = 9,
Because nums[0] + nums[1] = 2 + 7 = 9, return [0, 1]

```python
def twoSum(nlist, target):
    myDict = {}
    if not nlist:
        return None
    n = len(nlist)
    for index in range(0, n):
        num = nlist[index]
        if (target - num) in myDict:
            return [myDict[target - num], index+]
        else:
            myDict[num] = index 
    return None  
```

```python
twoSum( [2, 7, 11, 15],9)
```

# Leetcode 167. Two Sum II - Input array is sorted (Medium)

- Description:
Given an array of integers that is already sorted in ascending order,
find two numbers such that they add up to a specific target number.

The function twoSum should return indices of the two numbers such that they add up to the target,
where index1 must be less than index2.

Note:
Your returned answers (both index1 and index2) are not zero-based.
You may assume that each input would have exactly one solution and you may not use the same element twice.

- Example:
Input: numbers = [2,7,11,15], target = 9

Output: [1,2]

Explanation: The sum of 2 and 7 is 9. Therefore index1 = 1, index2 = 2.

```python
#use Two pointers
def twoSum2(nlist, target):
    start = 0
    end = len(nlist) - 1
    while start < end:
        if nlist[start] + nlist[end] == target:
            return [start+1, end+1]
        if nlist[start] + nlist[end] < target:
            start += 1
        else:
            end -= 1
    return None
```

```python
twoSum2([2,7,11,15], 9)
```

```python
# binary search
def binarySearch(nlist, target, start):
    left = start
    right = len(nlist) - 1
    while left <= right:
        mid = (left + right) // 2
        if nlist[mid] == target:
            return mid
        elif nlist[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return None
        
def twoSumBS(nlist, target):
    for index in range(len(nlist)):
        target_index = binarySearch(nlist, target - nlist[index], index+1)
        if target_index is not None:
            return [index+1, target_index+1]
    return []
```

```python
twoSumBS([2,7,11,15], 9)
```

# Leetcode 219. Contains Duplicate II

- Description:
Given an array of integers and an integer k,find out whether there are two distinct indices i and j
in the array such that nums[i] = nums[j] and the absolute difference between i and j is at most k.

- Example1:
Input: nums = [1,2,3,1], k = 3, Output: true

- Example2:
Input: nums = [1,0,1,1], k = 1, Output: true

- Example3:
Input: nums = [1,2,3,1,2,3], k = 2, Output: false


```python
def containDup(nlist, k):
    if not nlist:
        return False
    myDict = {}
    for index in range(len(nlist)):
        if nlist[index] in myDict and abs(index - myDict[nlist[index]]) <= k :
            return True
        myDict[nlist[index]] = index
    return False
    
```

```python
containDup([1,2,3,1], 3)
```

```python
containDup([1,0,1,1], 1)
```

```python
containDup([1,2,3,1,2,3], 2)
```

# Leetcode 242. Valid Anagram

- Description:
Given two strings s and t , write a function to determine if t is an anagram of s.

Note: You may assume the string contains only lowercase alphabets.

- Example1:
Input: s = "anagram", t = "nagaram", Output: true

- Example2:
Input: s = "rat", t = "car", Output: false


```python
def validAnagram(s, t):
    myDict1 = {}
    myDict2 = {}
    for i in range(len(s)):
        myDict1[s[i]] = myDict1.get(s[i], 0) + 1
    for i in range(len(t)):
        myDict2[t[i]] = myDict2.get(t[i], 0) + 1
    if myDict1 == myDict2:
        return True
    else:
        return False
```

```python
validAnagram("anagram", "nagaram")
```

```python
validAnagram("rat", "car")
```

# Leetcode 268. Missing Number

- Description:
Given an array containing n distinct numbers taken from 0, 1, 2, ..., n,
find the one that is missing from the array.

Note:
Your algorithm should run in linear runtime complexity. Could you implement it using only constant extra space complexity?

- Example1:
Input: [3,0,1], Output: 2

- Example2:
Input: [9,6,4,2,3,5,7,0,1], Output: 8


```python
# math
def missingNum(nlist):
    if not nlist:
        return None
    n = len(nlist) + 1
    return (n)*(n-1)//2 - sum(nlist)
    
```

```python
missingNum([3,0,1])
```

```python
missingNum([9,6,4,2,3,5,7,0,1])
```

```python
# dictionary
def missingNum2(nlist):
    if not nlist:
        return None
    
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in range(len(nlist)+1):
        if i not in myDict:
            return i
```

```python
missingNum2([3,0,1])
```

```python
missingNum([9,6,4,2,3,5,7,0,1])
```

# Leetcode 350. Intersection of Two Arrays II

- Description:
Given two arrays, write a function to compute their intersection.

Note:
Each element in the result should appear as many times as it shows in both arrays.
The result can be in any order.

- Example1:
Input: nums1 = [1,2,2,1], nums2 = [2,2], Output: [2,2]

- Example2:
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4], Output: [4,9]

```python
def intersectionTwoArrayII(nlist1, nlist2):
    myDict = {}
    nlist = []
    for i in nlist1:
        myDict[i] = myDict.get(i, 0) + 1
    for j in nlist2:
        if j in myDict:
            nlist.append(j)
            myDict[j] -= 1
            if myDict[j] == 0:
                del myDict[j]
    return nlist
```

```python
intersectionTwoArrayII([1,2,2,1], [2,2])
```

```python
intersectionTwoArrayII([4,9,5], [9,4,9,8,4])
```

# Leetcode 389. Find the Difference

- Description:
Given two strings s and t which consist of only lowercase letters.String t is generated by random shuffling string s and
then add one more letter at a random position.Find the letter that was added in t.

- Example 1:

s = "abcd"   t = "abcde", Output: e,                 Explanation: "e" is the letter that was added.

s = "abba" t="abbba", output: b



```python
def findDifference(s, t):
    myDict = {}
    for i in range(len(s)):
        myDict[s[i]] = myDict.get(s[i], 0) + 1
    for j in range(len(t)):
        if t[j] in myDict:
            myDict[t[j]] -= 1
            if myDict[t[j]] == 0:
                del myDict[t[j]]
        else:
            return t[j]
    return myDict.keys()
```

```python
findDifference("abcd", "abcde")
```

```python
findDifference("abba", "abbba")
```

# Leetcode 448. Find All Numbers Disappeared in an Array

- Description:

Given an array of integers where 1 ≤ a[i] ≤ n (n = size of array), some elements appear twice and others appear once.
Find all the elements of [1, n] inclusive that do not appear in this array.
Could you do it without extra space and in O(n) runtime? You may assume the returned list does not count as extra space.

- Example:
Input:[4,3,2,7,8,2,3,1] ,Output: [5,6]

```python
# dictionary
def findMissing(nlist):
    myDict = {}
    result = []
    n = len(nlist)
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for j in range(1, n):
        if j not in myDict:
            result.append(j)
    return result
```

```python
findMissing([4,3,2,7,8,2,3,1])
```

```python
def findMissing2(nlist):
    return list(set(range(1, len(nlist))) - set(nlist))
```

```python
def findMissing3(nlist):
    return list(set(range(1, len(nlist))) - set(nlist))
```

# Leetcode 28. Implement strstr(), Find the Index of the First Occurrence in a String 

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

- Example 1:

Input: haystack = "sadbutsad", needle = "sad"
Output: 0
Explanation: "sad" occurs at index 0 and 6.
The first occurrence is at index 0, so we return 0.
    
- Example 2:

Input: haystack = "leetcode", needle = "leeto"
Output: -1
Explanation: "leeto" did not occur in "leetcode", so we return -1

```python
def strstr2(haystack, needle):
    m = len(needle)
    n = len(haystack)
       
    if m == 0:
        return 0
    if n == 0:
        return -1
    if m > n:
        return -1
    for i in range(n-m+1):
        if haystack[i:i+m] == needle:
            return i
    return -1
```

```python
#brute force
#1. needle empty, haystack empty, return 0
def strstr(haystack, needle):
    m = len(needle)
    n = len(haystack)
       
    if m == 0:
        return 0
    if n == 0:
        return -1
    if m > n:
        return -1

    i = j = 0
    start = i
    while i < n and j < m:
        if needle[j] == haystack[i]:
            #print(i, j)
            i += 1
            j += 1
            if (j == m ) and (i < n+1):
                return start
        else:
            i += 1
            j = 0
            start = i
    return -1
            
```

# Leetcode 151. Reverse Words in a String (Medium)

Given an input string s, reverse the order of the words.

A word is defined as a sequence of non-space characters. The words in s will be separated by at least one space.

Return a string of the words in reverse order concatenated by a single space.

Note that s may contain leading or trailing spaces or multiple spaces between two words. The returned string should only have a single space separating the words. Do not include any extra spaces.

- Example 1:

Input: s = "the sky is blue"
Output: "blue is sky the"

- Example 2:

Input: s = "  hello world  "
Output: "world hello"
Explanation: Your reversed string should not contain leading or trailing spaces.

- Example 3:

Input: s = "a good   example"
Output: "example good a"
Explanation: You need to reduce multiple spaces between two words to a single space in the reversed string.

```python
def reverse_words(myStrings):
    if not myStrings:
        return []
    reversed_list =''
    n = len(myStrings)
    j = n
    for i in range(n-1, -1, -1):
        if myStrings[i] == ' ':
            j = i
        elif (i == 0 or myStrings[i-1] == ' '):
            if len(reversed_list) != 0:
                reversed_list +=' '
            reversed_list += (myStrings[i:j])
    return reversed_list   
```

```python
reverse_words("the sky is blue")
```

```python
reverse_words(" hello world")
```

# Leetcode 186. Reverse Words in a String II (Medium)

Given a character array s, reverse the order of the words.

A word is defined as a sequence of non-space characters. The words in s will be separated by a single space.

Your code must solve the problem in-place, i.e. without allocating extra space.

- Example 1:

Input: s = ["t","h","e"," ","s","k","y"," ","i","s"," ","b","l","u","e"]
Output: ["b","l","u","e"," ","i","s"," ","s","k","y"," ","t","h","e"]

- Example 2:

Input: s = ["a"]
Output: ["a"]

```python
def reverse2(slist, start, end):
    #reverse the list
    i = start
    j = end
    while i < j:
        slist[i], slist[j] = slist[j], slist[i]
        i += 1
        j -= 1
    return slist

def reverse_word2(slist):
    reverse2(slist, 0, len(slist)-1)
    j = 0
    for i in range(len(slist)):
        if slist[i] == ' ':
            reverse2(slist, j, i-1)
            j = i + 1
    reverse2(slist, j, i)
    return slist
```

# Leetcode 163. Missing Ranges (Medium)

Description
You are given an inclusive range [lower, upper] and a sorted unique integer array nums, where all elements are within the inclusive range.

A number x is considered missing if x is in the range [lower, upper] and x is not in nums.

Return the shortest sorted list of ranges that exactly covers all the missing numbers. That is, no element of nums is included in any of the ranges, and each missing number is covered by one of the ranges.

- Example 1:

Input: nums = [0,1,3,50,75], lower = 0, upper = 99
Output: [[2,2],[4,49],[51,74],[76,99]]

Explanation: The ranges are:
[2,2]
[4,49]
[51,74]
[76,99]

- Example 2:

Input: nums = [-1], lower = -1, upper = -1
Output: []

Explanation: There are no missing ranges since there are no missing numbers.

```python
def missingRange(nlist, lower, upper):
    i = lower
    j = upper
    n = len(nlist)
    result = []
    if n == 0:
        return [[lower, upper]]
    if lower < nlist[0]:
        result.append([lower, nlist[0] -1])
    k = 1
    while k <  n :
        if (nlist[k-1] + 1 ) < nlist[k]:
            result.append([nlist[k-1] + 1, nlist[k] - 1])
        k = k + 1
    
    if upper > nlist[n-1]:
        result.append([nlist[n-1]+1, upper])
    return result
```

```python
missingRange([-1], -1, -1)
```

```python
missingRange([], 1, 2)
```

# Leetcode 3. Longest Substring Without Repeating Characters (Medium)

Given a string s, find the length of the longest substring without duplicate characters.

- Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

- Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.

- Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.

```python
def longestSubstring(myStr):
    j=0
    result = 0
    myDict = {}
    for i, c in enumerate(myStr):
        if c in myDict:            
            myDict[c] = 1
            while c in myDict:
                myDict[myStr[j]] -= 1
                if myDict[myStr[j]] == 0:
                    del myDict[myStr[j]]
                j += 1
        myDict[c] = 1
        result = max(result, i-j+1)
    return result
```

```python
longestSubstring("abcabcbb")
```

```python
longestSubstring("bbbbb")
```

```python
longestSubstring("pwwkew")
```

```python
def longestTwoDistinct(myString):
    if not myString:
        return 0
    maxLength = 0

    for i in range(len(myString)):
        current = i
        charList = []
        #print('i:', i)
        while current < len(myString):  
            #print('current:', current)
            #print('charList:', charList)
            if myString[current] not in charList and len(charList)< 2:
                charList.append(myString[current])
                current += 1
            elif myString[current] not in charList and len(charList) >= 2:
                break
            else:
                current += 1
        length = current - i
        if length > maxLength:
            maxLength = length
    return maxLength    
            
```

```python
from collections import Counter
def lengthOfLongestSubstringTwoDistinct(s: str) -> int:
        cnt = Counter()
        ans = j = 0
        for i, c in enumerate(s):
            cnt[c] += 1
            while len(cnt) > 2:
                cnt[s[j]] -= 1
                if cnt[s[j]] == 0:
                    cnt.pop(s[j])
                j += 1
            ans = max(ans, i - j + 1)
        return ans
```

# Leetcode 65. Valid Number (Hard)

Given a string s, return whether s is a valid number.

For example, all the following are valid numbers: "2", "0089", "-0.1", "+3.14", "4.", "-.9", "2e10", "-90E3", "3e+7", "+6e-1", "53.5e93", "-123.456e789", while the following are not valid numbers: "abc", "1a", "1e", "e3", "99e2.5", "--6", "-+3", "95a54e53".

Formally, a valid number is defined using one of the following definitions:

An integer number followed by an optional exponent.
A decimal number followed by an optional exponent.
An integer number is defined with an optional sign '-' or '+' followed by digits.

A decimal number is defined with an optional sign '-' or '+' followed by one of the following definitions:

Digits followed by a dot '.'.
Digits followed by a dot '.' followed by digits.
A dot '.' followed by digits.
An exponent is defined with an exponent notation 'e' or 'E' followed by an integer number.

The digits are defined as one or more digits.

- Example 1:

Input: s = "0"

Output: true

- Example 2:

Input: s = "e"

Output: false

- Example 3:

Input: s = "."

Output: false
