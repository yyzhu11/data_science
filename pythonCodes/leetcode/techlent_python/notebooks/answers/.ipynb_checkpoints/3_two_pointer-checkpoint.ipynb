# Leetcode 159. Longest Substring with At Most Two Distinct Characters (Medium)

Description
Given a string s, return the length of the longest substring that contains at most two distinct characters.

- Example 1:

Input: s = "eceba"
Output: 3
Explanation: The substring is "ece" which its length is 3.

- Example 2:

Input: s = "ccaabbb"
Output: 5
Explanation: The substring is "aabbb" which its length is 5.

```python
def longestSubAtMostTwoChar(myStr):
    myDict = {}
    result = 0
    j = 0
    for i, c in enumerate(myStr):
        myDict[c] = myDict.get(c, 0) + 1
        while len(myDict) > 2:
            myDict[myStr[j]] -= 1
            if myDict[myStr[j]] == 0:
                del myDict[myStr[j]]
            j += 1
        result = max(result, i-j+1)
    return result
```

```python
longestSubAtMostTwoChar("eceba")
```

```python
longestSubAtMostTwoChar("ccaabbb")
```
