````python
#class node
from collections import deque
class TreeNode():
    def __init__(self, val=0, left=None, right=None):
        self.left = left
        self.right = right
        self.val = val

def to_binary_tree(items):
    if not items:
        return None
    it = iter(items)
    root = TreeNode(next(it))
    q = [root]
    for node in q:
        val = next(it, None)
        if val is not None:
            node.left = TreeNode(val)
            q.append(node.left)        
        val = next(it, None)
        if val is not None:
            node.right = TreeNode(val)
            q.append(node.right)
    return root

#Level Order Traversal
def print_binary_tree(root):
    queue = deque()
    result = []
    queue.append(root)
    while not all(item is None for item in queue):
        node = queue.popleft()
        if node:
            result.append(node.val)       
            if node.left is not None:
                #result.append(node.left.val)
                #print('add left:', node.left.val)
                queue.append(node.left)
            else:
                #print('add left: None')
                queue.append(None)
            if node.right is not None:
                #result.append(node.right.val)
                #print('add right:', node.right.val)
                queue.append(node.right)
            else:
                #print('add right: None')
                queue.append(None)
        else:
            result.append(None)
        
    return result     
````

## Preorder: recursive and stack

````python
def DFS_preorder(root):
    result = []
    def preorderTraversal(node):
        if not node:
            return         
        result.append(node.val)
        preorderTraversal(node.left)
        preorderTraversal(node.right)
    preorderTraversal(root)
    return result
````

````python
root = to_binary_tree([1,2,3,4,5,None, None,6])
DFS_preorder(root)
````

````python
def preorder_stack(root):
    if not root:
        return []
    stack = []
    result = []
    stack.append(root)
    while stack:
        node = stack.pop()
        result.append(node.val)
        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)
    return result
````

````python
root = to_binary_tree([1,2,3,4,5,None, None,6])
preorder_stack(root)
````

````python
def inorder_stack(root):
    if not root:
        return []
    stack = []
    result = []
    while stack or root:
        if root:
            stack.append(root)
            root = root.left
        else:
            node = stack.pop()
            result.append(node.val)
            root = node.right
    return result
````

## Postorder: recursive and stack, queue


- preorder : [1, 2, 4, 6, 5, 3]

- postorder: [6, 4, 5, 2, 3, 1]

````python
def DFS_postorder(root): 
    result = []
    def postorderTraversal(node):
        if not node:
            return   
        postorderTraversal(node.left)        
        postorderTraversal(node.right)
        result.append(node.val)
    postorderTraversal(root)
    return result
````

````python
root = to_binary_tree([1,2,3,4,5,None, None,6])
DFS_postorder(root)
````

````python
def postorder_stack(root):
    res = []
    stack = []
    cur = root
    while cur or stack:
        if cur:
            res.append(cur.val)
            stack.append(cur)
            cur = cur.right
        else:
            cur = stack.pop()
            cur = cur.left

    res.reverse()
    return res
````

````python
root = to_binary_tree([1,2,3,4,5,None, None,6])
postorder_stack(root)
````

````python
def postorder_queue(root):
    result = deque()
    def BFS(root):
        queue = deque([root])        
        while queue:
            node = queue.popleft()
            result.appendleft(node.val)
            if node.left:
                queue.appendleft(node.left)
            if node.right:
                queue.appendleft(node.right)
                
    BFS(root)
    return list(result)
````

````python
root = to_binary_tree([1,2,3,4,5,None, None,6])
postorder_queue(root)
````

## Leetcode 144. Binary Tree Preorder Traversal

Description:
Given a binary tree, return the preorder traversal of its nodes’ values.

The function twoSum should return indices of the two numbers such that they add up to the target, where index1 must be less than index2.

Example:
Input: [1,null,2,3]
```
   1
    \
     2
    /
   3
```
Output: [1,2,3]

````python
def preorderTraversal(root):
    """
    :type root: Optional[TreeNode]
    :rtype: List[int]
    """
    result = []
    def traversal(node):
        if not node:
            return 
        result.append(node.val)
        traversal(node.left)
        traversal(node.right)
    
    traversal(root)
    return result
````

## Leetcode 257. Binary Tree Paths

Description:
Given a binary tree, return all root-to-leaf paths.

Note:
A leaf is a node with no children.

- Example 1:
Input:
```
      1
    /   \
   2     3
    \
     5
```
Output: ["1->2->5", "1->3"]

- Example 2:

Input: root = [1]
Output: ["1"]


````python
def getAllPath(root):
    def preorder(node):
        if node is None:
            return 
        
        current_result.append(str(node.val))
        if not node.left and not node.right:
            all_result.append('->'.join(current_result))
        else:
            preorder(node.left)
            preorder(node.right)       
        # Backtrack: remove current node from path before returning
        current_result.pop()
        
    all_result=[]
    current_result = []
    preorder(root)
    return all_result   
````

Description:
Given a binary tree and a sum, determine if the tree has a root-to-leaf path such that adding up all the values along the path equals the given sum.

Note:
A leaf is a node with no children.
- Example 1:
Given the below binary tree and sum = 22,
```
      5
    / \
   4   8
  /   / \
 11  13  4
/  \      \
7    2      1
```
return true, as there exist a root-to-leaf path 5->4->11->2 which sum is 22.


- Example 2:

```
        1
       / \
      2   3
```


Input: root = [1,2,3], targetSum = 5
Output: false

Explanation: There are two root-to-leaf paths in the tree:
(1 --> 2): The sum is 3.
(1 --> 3): The sum is 4.
There is no root-to-leaf path with sum = 5.

````python
def pathSum(root, num):
    result = 0
    all_result = []
    def preorder(node, current_sum):
        if not node:
            return False
        current_sum += node.val
        if not node.left and not node.right:
            if current_sum == num:
                return True
        return preorder(node.left, current_sum) or preorder(node.right, current_sum)
        
    return preorder(root, 0)
````

## Leetcode 113. Path Sum II (Medium)

Given the root of a binary tree and an integer targetSum, return all root-to-leaf paths where the sum of the node values in the path equals targetSum. Each path should be returned as a list of the node values, not node references.

A root-to-leaf path is a path starting from the root and ending at any leaf node. A leaf is a node with no children.

- Example 1:



Input: root = [5,4,8,11,null,13,4,7,2,null,null,5,1], targetSum = 22
Output: [[5,4,11,2],[5,8,4,5]]
Explanation: There are two paths whose sum equals targetSum:
5 + 4 + 11 + 2 = 22
5 + 8 + 4 + 5 = 22

- Example 2:



Input: root = [1,2,3], targetSum = 5
Output: []

- Example 3:

Input: root = [1,2], targetSum = 0
Output: []

````python
def pathSumII(root, targetSum):

    def preorder(node):
        if node is None:
            return
        
        current_result.append(node.val)
        if not node.left and not node.right:
            #print(current_result)
            current_sum = sum(current_result)
            if current_sum == targetSum: 
                #Otherwise, all_result will end up with empty lists after popping
                #In Python, lists are objects. If you append current_result directly, 
                #you are appending a "link" to that list.
                all_result.append(list(current_result)) 
        else:
            preorder(node.left)
            preorder(node.right)
        current_result.pop()

    all_result = []
    current_result = []
    preorder(root)
    return all_result
````

## Leetcode 515. Find Largest Value in Each Tree Row (Medium)

Description:
Given the root of a binary tree, return an array of the largest value in each row of the tree (0-indexed).

- Example 1:
```
        1
       / \
      3   2
    /  \   \
    5   3   9

```
Input: root = [1,3,2,5,3,null,9]

Output: [1,3,9]

- Example 2:
  
Input: root = [1,2,3]

Output: [1,3]

- Example 3:

Input: root = [1]

Output: [1]


````python
def maxRow(root):
    final = []
    def preorderTraversal(node, height):
        if not node:
            return
        if height == len(final):
            final.append(node.val)
        final[height]= max(final[height], node.val)
        preorderTraversal(node.left, height+1)
        preorderTraversal(node.right, height+1)
    
    preorderTraversal(root, 0)
    return final
````

Given the root of a binary search tree, and an integer k, return the kth smallest value (1-indexed) of all the values of the nodes in the tree.

Input: root = [3,1,4,null,2], k = 1
Output: 1
- Example 2:



Input: root = [5,3,6,2,4,null,null,1], k = 3
Output: 3

````python
def kthSmallestBT(root, k):
    result = []
    def inorder(root):
        if root is None:
            return []
        inorder(root.left)
        result.append(root.val)
        inorder(root.right)

    inorder(root)
    return result[k-1]
````

## Leetcode 98. Validate Binary Search Tree (Medium)

Given the root of a binary tree, determine if it is a valid binary search tree (BST).

A valid BST is defined as follows:

The left subtree of a node contains only nodes with keys strictly less than the node's key.
The right subtree of a node contains only nodes with keys strictly greater than the node's key.
Both the left and right subtrees must also be binary search trees.

- Example 1:
```
   2
  /  \
 1   3
```
Input: root = [2,1,3]
Output: true

- Example 2:
```
     5
    /  \
   1     4
        /  \
        3   6
```
Input: root = [5,1,4,null,null,3,6]
Output: false

Explanation: The root node's value is 5 but its right child's value is 4.

- Example 3:

```
     5
    /   \
   3     8
  / \      \
 2   4      9
```
Input: root = [5,3,8,2,4,None,9]
Output: false

- Example 4:

```
     5
    /   \
   3     8
  / \      \
 2   6      9
```
Input: root = [5,3,8,2,6,None,9]
Output: false


````python
import math
def validateBT(root):
    def inorder_traversal(node):
        if node is None:
            return True
        if not inorder_traversal(node.left):
            return False
        nonlocal pre_val
        if pre_val >= node.val:
            return False
        pre_val = node.val
        return inorder_traversal(node.right)

    pre_val = -math.inf
    return inorder_traversal(root)
````

## Leetcode 99. Recover Binary Search Tree (Medium)

You are given the root of a binary search tree (BST), where the values of exactly two nodes of the tree were swapped by mistake. Recover the tree without changing its structure.

- Example 1:



Input: root = [1,3,null,null,2]
Output: [3,1,null,null,2]

Explanation: 3 cannot be a left child of 1 because 3 > 1. Swapping 1 and 3 makes the BST valid.

- Example 2:



Input: root = [3,1,4,null,null,2]
Output: [2,1,4,null,null,3]

Explanation: 2 cannot be in the right subtree of 3 because 2 < 3. Swapping 2 and 3 makes the BST valid.

````python
def recoverBT(root):
    def inorder_traversal(node):
        if node is None:
            return None, None
        inorder_traversal(node.left)

        nonlocal pre_val
        nonlocal pre_node
        if pre_val >= node.val:
            #these nodes need to swap
            print(pre_val, node.val)
            return pre_node, node
            
        pre_val = node.val
        pre_node = node
        inorder_traversal(node.right)

    pre_val = -math.inf
    pre_node = None

    node1, node2 = inorder_traversal(root)
    print('node1:', node1.val)
    print('node2:', node2.val)
````

## Leetcode 637. Average of Levels in Binary Tree

Given the root of a binary tree, return the average value of the nodes on each level in the form of an array. Answers within 10-5 of the actual answer will be accepted.

- Example 1:



Input: root = [3,9,20,null,null,15,7]
Output: [3.00000,14.50000,11.00000]
Explanation: The average value of nodes on level 0 is 3, on level 1 is 14.5, and on level 2 is 11.
Hence return [3, 14.5, 11].

- Example 2:



Input: root = [3,9,20,15,7]
Output: [3.00000,14.50000,11.00000]

````python
def avgBTLevel(root):
    result = []
    queue = deque()
    if not root:
        return []
    queue.append(root)
    while queue:
        num = len(queue)
        level_sum = 0
        for i in range (num):
            node = queue.popleft()
            level_sum += node.val
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        mean = level_sum / num
        result.append(mean)
    return result
````

## Leetcode 102. Binary Tree Level Order Traversal (Medium)

Given the root of a binary tree, return the level order traversal of its nodes' values. (i.e., from left to right, level by level).

- Example 1:

```
         3
        / \
       9   20
           / \
          15  7


```

Input: root = [3,9,20,null,null,15,7]
Output: [[3],[9,20],[15,7]]

- Example 2:

Input: root = [1]
Output: [[1]]

- Example 3:

Input: root = []
Output: []

````python
def BT_level_order(root):
    result = []       
    if root is None:
        return []
    
    queue = deque()
    queue.append(root)    
    
    while queue:
        num = len(queue)
        level_result = []
        for i in range(num):
            item = queue.popleft()
            level_result.append(item.val)
            if item.left:
                queue.append(item.left)
            if item.right:
                queue.append(item.right)
        result.append(level_result)
    return result
    
````

````python
root = to_binary_tree( [3,9,20,None,None,15,7])
BT_level_order(root)
````

````python
root = to_binary_tree( [1])
BT_level_order(root)
````

````python
root = to_binary_tree([])
BT_level_order(root)
````

## Leetcode 103. Binary Tree Zigzag Level Order Traversal (Medium)
Medium

Given the root of a binary tree, return the zigzag level order traversal of its nodes' values. (i.e., from left to right, then right to left for the next level and alternate between).

- Example 1:



Input: root = [3,9,20,null,null,15,7]
Output: [[3],[20,9],[15,7]]

- Example 2:

Input: root = [1]
Output: [[1]]
- Example 3:

Input: root = []
Output: []

````python
def BT_level_order_zigzag(root):
    if root is None:
        return []
    result = []
    queue = deque()
    queue.append(root)
    zigzag = 0
    while queue:
        num = len(queue)
        zigzag = 1 - zigzag #left
        level_result = []
        for i in range(num):
            node = queue.popleft()
            level_result.append(node.val)
            if zigzag == 0:
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            else:
                if node.right:
                    queue.append(node.right)
                if node.left:
                    queue.append(node.left)
        
        result.append(level_result)
    return result
        
````

````python
root = to_binary_tree([3,9,20,None,None,15,7])
BT_level_order_zigzag(root)
````

## Leetcode 515. Find Largest Value in Each Tree Row (Medium)

Description:
Given the root of a binary tree, return an array of the largest value in each row of the tree (0-indexed).

- Example 1:
```
        1
       / \
      3   2
    /  \   \
    5   3   9

```
Input: root = [1,3,2,5,3,null,9]

Output: [1,3,9]

- Example 2:
  
Input: root = [1,2,3]

Output: [1,3]

- Example 3:

Input: root = [1]

Output: [1]


````python
def maxRow(root):
    result = []
    if root is None:
        return []
    queue = deque()
    queue.append(root)
    while queue:
        num = len(queue)
        level_result = []
        for i in range(num):
            node = queue.popleft()
            level_result.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        level_max = max(level_result)
        result.append(level_max)
    return result
````

````python
root = to_binary_tree([1,3,2,5,3,None,9])
maxRow(root)
````

````python
root = to_binary_tree([1,2,3])
maxRow(root)
````

## Leetcode 199. Binary Tree Right Side View

Description:
Given a binary tree, imagine yourself standing on the right side of it, return the values of the nodes you can see ordered from top to bottom.

- Example:
Input: [1,2,3,null,5,null,4]

Output: [1, 3, 4]

Explanation:
```
      1            <---
    /   \
   2     3         <---
    \     \
     5     4       <---
```

- Example 2:

Input: root = [1,2,3,4,null,null,null,5]

Output: [1,3,4,5]

```
            1
           / \
          2   3
         /
        4
       /
      5
        
```

Explanation:

````python
def BTRightSideView(root):
    level_result = []
    final_result =[]
    if root is None:
        return None
    queue = deque()
    queue.append(root)
    while queue:
        num = len(queue)
        level_result = []
        for i in range(num):
            node = queue.popleft()
            level_result.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        final_result.append(level_result[-1])
    
    return final_result
````
