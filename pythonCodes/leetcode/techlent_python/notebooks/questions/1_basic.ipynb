# leetcode 136. Single Number

Given a non-empty array of integers, every element appears twice except for one. Find that single one.

Note:
Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

- Example: Input: [2,2,1], Output: 1


```python
# use math
def singleNumber(nlist):
    mySet = set(nlist)
    return sum(mySet)*2 - sum(nlist)
```

```python
# use dictionary
def singleNumber2(nlist):
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict.keys():
        if myDict[i] == 1:
            return i
```

# leetcode 169. Majority Element

Description:
Given an array of size n, find the majority element. The majority element is the element that appears more than ⌊ n/2 ⌋ times.

Note:
You may assume that the array is non-empty and the majority element always exist in the array.

- Example:
  Input: [3,2,3], Output: 3

```python
def majorityElement(nlist):
    if not nlist:
        return None
    n = len(nlist)
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict.keys():
        if myDict[i] > n//2:
            return i
```

```python
majorityElement([3,2,3])
```

# leetcode 217. Contains Duplicate

- Description:
Given an array of integers, find if the array contains any duplicates.

Note:
Your function should return true if any value appears at least twice in the array, and it should return false
if every element is distinct.

- Example: Input: [1,2,3,1], Output: true

```python
def containDuplicate(nlist):
  if not nlist:
    return False
  myDict = {}
  for i in nlist:
    if i in myDict:
      return True
    else:
      myDict[i] = myDict.get(i, 0)+1
  return False
```

# Leetcode 1. Two Sum

- Description:
Given an array of integers, return indices of the two numbers such that they add up to a specific target.

Note:
You may assume that each input would have exactly one solution, and you may not use the same element twice.

- Example:
Given nums = [2, 7, 11, 15], target = 9,
Because nums[0] + nums[1] = 2 + 7 = 9, return [0, 1]

```python
def twoSum(nlist, target):
    myDict = {}
    if not nlist:
        return None
    n = len(nlist)
    for index in range(0, n):
        num = nlist[index]
        if (target - num) in myDict:
            return [myDict[target - num], index+]
        else:
            myDict[num] = index 
    return None  
```

```python
twoSum( [2, 7, 11, 15],9)
```

# Leetcode 167. Two Sum II - Input array is sorted

- Description:
Given an array of integers that is already sorted in ascending order,
find two numbers such that they add up to a specific target number.

The function twoSum should return indices of the two numbers such that they add up to the target,
where index1 must be less than index2.

Note:
Your returned answers (both index1 and index2) are not zero-based.
You may assume that each input would have exactly one solution and you may not use the same element twice.

- Example:
Input: numbers = [2,7,11,15], target = 9

Output: [1,2]

Explanation: The sum of 2 and 7 is 9. Therefore index1 = 1, index2 = 2.

```python
#use Two pointers
def twoSum2(nlist, target):
    start = 0
    end = len(nlist) - 1
    while start < end:
        if nlist[start] + nlist[end] == target:
            return [start+1, end+1]
        if nlist[start] + nlist[end] < target:
            start += 1
        else:
            end -= 1
    return None
```

```python
twoSum2([2,7,11,15], 9)
```

```python
# binary search
def binarySearch(nlist, target, start):
    left = start
    right = len(nlist) - 1
    while left <= right:
        mid = (left + right) // 2
        if nlist[mid] == target:
            return mid
        elif nlist[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return None
        
def twoSumBS(nlist, target):
    for index in range(len(nlist)):
        target_index = binarySearch(nlist, target - nlist[index], index+1)
        if target_index is not None:
            return [index+1, target_index+1]
    return []
```

# Leetcode 219. Contains Duplicate II

- Description:
Given an array of integers and an integer k,find out whether there are two distinct indices i and j
in the array such that nums[i] = nums[j] and the absolute difference between i and j is at most k.

- Example1:
Input: nums = [1,2,3,1], k = 3, Output: true

- Example2:
Input: nums = [1,0,1,1], k = 1, Output: true

- Example3:
Input: nums = [1,2,3,1,2,3], k = 2, Output: false


```python
def containDup(nlist, k):
    if not nlist:
        return False
    myDict = {}
    for index in range(len(nlist)):
        if nlist[index] in myDict and abs(index - myDict[nlist[index]]) <= k :
            return True
        myDict[nlist[index]] = index
    return False
    
```

```python
containDup([1,2,3,1], 3)
```

```python
containDup([1,0,1,1], 1)
```

```python
containDup([1,2,3,1,2,3], 2)
```

# Leetcode 242. Valid Anagram

- Description:
Given two strings s and t , write a function to determine if t is an anagram of s.

Note: You may assume the string contains only lowercase alphabets.

- Example1:
Input: s = "anagram", t = "nagaram", Output: true

- Example2:
Input: s = "rat", t = "car", Output: false


```python
def validAnagram(s, t):
    myDict1 = {}
    myDict2 = {}
    for i in range(len(s)):
        myDict1[s[i]] = myDict1.get(s[i], 0) + 1
    for i in range(len(t)):
        myDict2[t[i]] = myDict2.get(t[i], 0) + 1
    if myDict1 == myDict2:
        return True
    else:
        return False
```

```python
validAnagram("anagram", "nagaram")
```

```python
validAnagram("rat", "car")
```

# Leetcode 268. Missing Number

- Description:
Given an array containing n distinct numbers taken from 0, 1, 2, ..., n,
find the one that is missing from the array.

Note:
Your algorithm should run in linear runtime complexity. Could you implement it using only constant extra space complexity?

- Example1:
Input: [3,0,1], Output: 2

- Example2:
Input: [9,6,4,2,3,5,7,0,1], Output: 8


```python
# math
def missingNum(nlist):
    if not nlist:
        return None
    n = len(nlist) + 1
    return (n)*(n-1)//2 - sum(nlist)
    
```

```python
missingNum([3,0,1])
```

```python
missingNum([9,6,4,2,3,5,7,0,1])
```

```python
# dictionary
def missingNum2(nlist):
    if not nlist:
        return None
    
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in range(len(nlist)+1):
        if i not in myDict:
            return i
```

```python
missingNum2([3,0,1])
```

```python
missingNum([9,6,4,2,3,5,7,0,1])
```

# Leetcode 350. Intersection of Two Arrays II

- Description:
Given two arrays, write a function to compute their intersection.

Note:
Each element in the result should appear as many times as it shows in both arrays.
The result can be in any order.

- Example1:
Input: nums1 = [1,2,2,1], nums2 = [2,2], Output: [2,2]

- Example2:
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4], Output: [4,9]

```python
def intersectionTwoArrayII(nlist1, nlist2):
    myDict = {}
    nlist = []
    for i in nlist1:
        myDict[i] = myDict.get(i, 0) + 1
    for j in nlist2:
        if j in myDict:
            nlist.append(j)
            myDict[j] -= 1
            if myDict[j] == 0:
                del myDict[j]
    return nlist
```

```python
intersectionTwoArrayII([1,2,2,1], [2,2])
```

```python
intersectionTwoArrayII([4,9,5], [9,4,9,8,4])
```

# Leetcode 389. Find the Difference

- Description:
Given two strings s and t which consist of only lowercase letters.String t is generated by random shuffling string s and
then add one more letter at a random position.Find the letter that was added in t.

- Example 1:

s = "abcd"   t = "abcde", Output: e,                 Explanation: "e" is the letter that was added.

s = "abba" t="abbba", output: b



```python
def findDifference(s, t):
    myDict = {}
    for i in range(len(s)):
        myDict[s[i]] = myDict.get(s[i], 0) + 1
    for j in range(len(t)):
        if t[j] in myDict:
            myDict[t[j]] -= 1
            if myDict[t[j]] == 0:
                del myDict[t[j]]
        else:
            return t[j]
    return myDict.keys()
```

```python
findDifference("abcd", "abcde")
```

```python
findDifference("abba", "abbba")
```

# Leetcode 448. Find All Numbers Disappeared in an Array

- Description:

Given an array of integers where 1 ≤ a[i] ≤ n (n = size of array), some elements appear twice and others appear once.
Find all the elements of [1, n] inclusive that do not appear in this array.
Could you do it without extra space and in O(n) runtime? You may assume the returned list does not count as extra space.

- Example:
Input:[4,3,2,7,8,2,3,1] ,Output: [5,6]

```python
# dictionary
def findMissing(nlist):
    myDict = {}
    result = []
    n = len(nlist)
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for j in range(1, n):
        if j not in myDict:
            result.append(j)
    return result
```

```python
findMissing([4,3,2,7,8,2,3,1])
```

```python
def findMissing2(nlist):
    return list(set(range(1, len(nlist))) - set(nlist))
```

```python
findMissing2([4,3,2,7,8,2,3,1])
```

```python
def findMissing3(nlist):
    return list(set(range(1, len(nlist))) - set(nlist))
```

- Description:
You have n candies, the ith candy is of type candies[i].

You want to distribute the candies equally between a sister and a brother so that each of them gets n / 2 candies (n is even).

The sister loves to collect different types of candies, so you want togive her the maximum number of different types of candies.

Return the maximum number of different types of candies you can give to the sister.

- Example1:
Input: candies =  [1,1,2,3] , Output: 2

The sister has candies [1,2,3]

```python
def distributeCandies(nlist):
    n = len(nlist) // 2
    
    return min(len(set(nlist)), n)
```
