# leetcode 136. Single Number

Given a non-empty array of integers, every element appears twice except for one. Find that single one.

Note:
Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

- Example 1:

Input: nums = [2,2,1]

Output: 1

- Example 2:

Input: nums = [4,1,2,1,2]

Output: 4

- Example 3:

Input: nums = [1]

Output: 1

```python
import sys
print sys.__python__
```

```python
def singleNumber(nlist):
    if not nlist:
        return None
    result = nlist[0]
    for i in nlist[1:]:
        result = result ^ i
    return result
```

```python
singleNumber([3,3,5])
```

```python
singleNumber([4,1,2,1,2])
```

```python
def singleNumber2(nlist):
    if not nlist:
        return None
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict:
        if myDict[i] == 1:
            return i
    return None
```

```python
singleNumber([2,2,3])
```

```python
def singleNumber3(nlist):
    if not nlist:
        return None
    return sum(set(nlist))*2 - sum(nlist)
```

```python
singleNumber([2,2,1])
```

```python
singleNumber([2,2,3])
```

# leetcode 169. Majority Element

Description:
Given an array of size n, find the majority element. The majority element is the element that appears more than ⌊ n/2 ⌋ times.

Note:
You may assume that the array is non-empty and the majority element always exist in the array.

- Example 1:

Input: nums = [3,2,3]
Output: 3

- Example 2:

Input: nums = [2,2,1,1,1,2,2]
Output: 2

```python
# O(n), m(n)
def majorityElement(nlist):
    if not nlist:
        return None
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for j in myDict:
        if myDict[j] > (len(nlist)//2):
            return j
    return None   
```

```python
# O(nlogn)
def majorityElement2(nlist):
    nlist.sort()
    return nlist[len(nlist)//2]
```

```python
#Boyer-Moore Majority Voting Algorithm
# O(n), m(1)
def majorityElement3(nlist):
    cnt = m = 0
    for x in nlist:        
        if cnt == 0:
            m, cnt = x, 1
        else:
            if m == x:
                cnt += 1 
            else: 
                cnt -= 1
    return m
```

```python
majorityElement3([2,2,1,1,1,2,2])
```

```python
majorityElement2([3,2,3])
```

```python
majorityElement2([2,2,1,1,1,2,2])
```

# leetcode 217. Contains Duplicate

- Description:
Given an array of integers, find if the array contains any duplicates.

Note:
Your function should return true if any value appears at least twice in the array, and it should return false
if every element is distinct.

- Example 1:

Input: nums = [1,2,3,1]

Output: true

Explanation:The element 1 occurs at the indices 0 and 3.

- Example 2:

Input: nums = [1,2,3,4]

Output: false

Explanation:All elements are distinct.

- Example 3:

Input: nums = [1,1,1,3,3,4,3,2,4,2]

Output: true

```python
# O(n), m(n)
def containDuplicate(nlist):
    if not nlist:
        return False
    myDict = {}
    for i in nlist:
        if i in myDict:
            return True
        else:
            myDict[i] = 1
    return False
```

# Leetcode 219. Contains Duplicate II

- Description:
Given an array of integers and an integer k,find out whether there are two distinct indices i and j
in the array such that nums[i] = nums[j] and the absolute difference between i and j is at most k.

- Example1:
Input: nums = [1,2,3,1], k = 3, Output: true

- Example2:
Input: nums = [1,0,1,1], k = 1, Output: true

- Example3:
Input: nums = [1,2,3,1,2,3], k = 2, Output: false


```python
def containsDup2(nlist, k):
	myDict = {}
	n = len(nlist)
	for i in range(n):
		if nlist[i] in myDict and abs(i - myDict[nlist[i]]) <= k:
			return True
		else:
			myDict[nlist[i]]=i
	return False
```

```python
containsDup2( [1,2,3,1], 3)
```

```python
containsDup2( [1,0,1,1], 1)
```

```python
containsDup2( [1,2,3,1,2,3], 2)
```

```python
def containDup(nlist, k):
    if not nlist:
        return False
    myDict = {}
    for index in range(len(nlist)):
        if nlist[index] in myDict and abs(index - myDict[nlist[index]]) <= k :
            return True
        myDict[nlist[index]] = index
    return False
    
```

# Leetcode 1. Two Sum

- Description:
Given an array of integers, return indices of the two numbers such that they add up to a specific target.

Note:
You may assume that each input would have exactly one solution, and you may not use the same element twice.

- Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

- Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]

- Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]

```python
def twoSum(nlist, target):
	myDict = {}
	n = len(nlist)
	for i in range(n):
		if (target - nlist[i]) in myDict:
			return [myDict[(target - nlist[i])], i]
		myDict[nlist[i]] = i
	return []
```

```python
twoSum([2,7,11,15], 9)
```

```python
twoSum([3,2,4], 6)
```

```python
twoSum([3,3], 6)
```

```python
def twoSum(nlist, target):
    myDict = {}
    if not nlist:
        return None
    n = len(nlist)
    for index in range(0, n):
        num = nlist[index]
        if (target - num) in myDict:
            return [myDict[target - num], index+]
        else:
            myDict[num] = index 
    return None  
```

```python
twoSum([2, 7, 11, 15],9)
```

# Leetcode 167. Two Sum II - Input array is sorted (Medium)

- Description:
Given an array of integers that is already sorted in ascending order,
find two numbers such that they add up to a specific target number.

The function twoSum should return indices of the two numbers such that they add up to the target,
where index1 must be less than index2.

Note:
Your returned answers (both index1 and index2) are not zero-based.
You may assume that each input would have exactly one solution and you may not use the same element twice.

- Example:
Input: numbers = [2,7,11,15], target = 9

Output: [1,2]

Explanation: The sum of 2 and 7 is 9. Therefore index1 = 1, index2 = 2.

- Example 2:

Input: numbers = [2,3,4], target = 6
Output: [1,3]
Explanation: The sum of 2 and 4 is 6. Therefore index1 = 1, index2 = 3. We return [1, 3].

- Example 3:

Input: numbers = [-1,0], target = -1
Output: [1,2]
Explanation: The sum of -1 and 0 is -1. Therefore index1 = 1, index2 = 2. We return [1, 2].

```python
#use two pointer
def twoSumII(nlist, target):
    left, right = 0, len(nlist)-1
    while left < right:
        if (nlist[left] + nlist[right]) == target:
            return [left+1, right+1]
        elif (nlist[left] + nlist[right]) > target:
            right -= 1
        else:
            left += 1
    return []
```

```python
twoSumII([2,7,11,15],9)
```

```python
twoSumII([2,3,4],6)
```

```python
twoSumII([-1,0], -1)
```

```python
#use Two pointers
def twoSum2(nlist, target):
    start = 0
    end = len(nlist) - 1
    while start < end:
        if nlist[start] + nlist[end] == target:
            return [start+1, end+1]
        if nlist[start] + nlist[end] < target:
            start += 1
        else:
            end -= 1
    return None
```

```python
twoSum2([2,7,11,15], 9)
```

```python
# binary search
def binarySearch(nlist, target, start):
    left = start
    right = len(nlist) - 1
    while left <= right:
        mid = (left + right) // 2
        if nlist[mid] == target:
            return mid
        elif nlist[mid] > target:
            right = mid - 1
        else:
            left = mid + 1
    return None
        
def twoSumBS(nlist, target):
    for index in range(len(nlist)):
        target_index = binarySearch(nlist, target - nlist[index], index+1)
        if target_index is not None:
            return [index+1, target_index+1]
    return []
```

```python
twoSumBS([2,7,11,15], 9)
```

# Leetcode 242. Valid Anagram

- Description:
Given two strings s and t , write a function to determine if t is an anagram of s.

Note: You may assume the string contains only lowercase alphabets.

- Example1:
Input: s = "anagram", t = "nagaram", Output: true

- Example2:
Input: s = "rat", t = "car", Output: false


```python
from collections import Counter
```

```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False
    sCounter = Counter(s)
    tCounter = Counter(t)
    return sCounter == tCounter 
    
```

```python
isAnagram("anagram", "nagaram")
```

```python
isAnagram("rat", "car")
```

```python
def validAnagram(s, t):
    myDict1 = {}
    myDict2 = {}
    for i in range(len(s)):
        myDict1[s[i]] = myDict1.get(s[i], 0) + 1
    for i in range(len(t)):
        myDict2[t[i]] = myDict2.get(t[i], 0) + 1
    if myDict1 == myDict2:
        return True
    else:
        return False
```

```python
validAnagram("anagram", "nagaram")
```

```python
validAnagram("rat", "car")
```

# Leetcode 350. Intersection of Two Arrays II

- Description:
Given two arrays, write a function to compute their intersection.

Note:
Each element in the result should appear as many times as it shows in both arrays.
The result can be in any order.

- Example1:
Input: nums1 = [1,2,2,1], nums2 = [2,2], Output: [2,2]

- Example2:
Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4], Output: [4,9]

```python
def intersectionTwoArrayII(nlist1, nlist2):
    myDict = {}
    nlist = []
    for i in nlist1:
        myDict[i] = myDict.get(i, 0) + 1
    for j in nlist2:
        if j in myDict:
            nlist.append(j)
            myDict[j] -= 1
            if myDict[j] == 0:
                del myDict[j]
    return nlist
```

```python
intersectionTwoArrayII([1,2,2,1], [2,2])
```

```python
intersectionTwoArrayII([4,9,5], [9,4,9,8,4])
```

# Leetcode 389. Find the Difference

- Description:
Given two strings s and t which consist of only lowercase letters.String t is generated by random shuffling string s and
then add one more letter at a random position.Find the letter that was added in t.

- Example 1:

s = "abcd"   t = "abcde", Output: e,                 Explanation: "e" is the letter that was added.

s = "abba" t="abbba", output: b



```python
def findDifference(s, t):
    myDict = {}
    for i in range(len(s)):
        myDict[s[i]] = myDict.get(s[i], 0) + 1
    for j in range(len(t)):
        if t[j] in myDict:
            myDict[t[j]] -= 1
            if myDict[t[j]] == 0:
                del myDict[t[j]]
        else:
            return t[j]
    return myDict.keys()
```

```python
findDifference("abcd", "abcde")
```

```python
findDifference("abba", "abbba")
```

# Leetcode 205. Isomorphic Strings
Given two strings s and t, determine if they are isomorphic.

Two strings s and t are isomorphic if the characters in s can be replaced to get t.

All occurrences of a character must be replaced with another character while preserving the order of characters. No two characters may map to the same character, but a character may map to itself.

- Example 1:

Input: s = "egg", t = "add", Output: true

Explanation:
The strings s and t can be made identical by:

Mapping 'e' to 'a'.
Mapping 'g' to 'd'.

- Example 2:

Input: s = "f11", t = "b23", Output: false

Explanation:

The strings s and t can not be made identical as '1' needs to be mapped to both '2' and '3'.

- Example 3:

Input: s = "paper", t = "title"

Output: true

```python
def isIsomorphic(s, t):
    myDict = {}
    if len(s) != len(t):
        return False
    n = len(s)
    for i in range(n):
        if s[i] not in myDict:
            myDict[s[i]] = t[i]
        else:
            if myDict[s[i]] != t[i]:
                return False
    return True
    
```

```python
isIsomorphic("egg", "add")
```

```python
isIsomorphic("f11", "b23")
```

```python
isIsomorphic("paper", "title")
```

# Leetcode 290. Word Pattern
Given a pattern and a string s, find if s follows the same pattern.

Here follow means a full match, such that there is a bijection between a letter in pattern and a non-empty word in s. Specifically:

Each letter in pattern maps to exactly one unique word in s.
Each unique word in s maps to exactly one letter in pattern.
No two letters map to the same word, and no two words map to the same letter.

- Example 1:
Input: pattern = "abba", s = "dog cat cat dog", Output: true

Explanation:

The bijection can be established as:

'a' maps to "dog".
'b' maps to "cat".
- Example 2:

Input: pattern = "abba", s = "dog cat cat fish", Output: false

- Example 3:

Input: pattern = "aaaa", s = "dog cat cat dog" , Output: false

```python
def wordPattern(s, t):
    words = t.split(' ')
    myDict = {}
    if len(s) != len(words):
        return False
    for i in range(len(s)):
        if s[i] not in myDict:
            myDict[s[i]] = words[i]
        else:
            if myDict[s[i]] != words[i]:
                return False
    return True
```

```python
wordPattern("abba", "dog cat cat dog")
```

```python
wordPattern("abba", "dog cat cat fish")
```

```python
wordPattern("aaaa", "dog cat cat dog")
```

Given two strings needle and haystack, return the index of the first occurrence of needle in haystack, or -1 if needle is not part of haystack.

- Example 1:

Input: haystack = "sadbutsad", needle = "sad"
Output: 0
Explanation: "sad" occurs at index 0 and 6.
The first occurrence is at index 0, so we return 0.

- Example 2:

Input: haystack = "leetcode", needle = "leeto"
Output: -1
Explanation: "leeto" did not occur in "leetcode", so we return -1

```python
def strstr2(haystack, needle):
    m = len(needle)
    n = len(haystack)
       
    if m == 0:
        return 0
    if n == 0:
        return -1
    if m > n:
        return -1
    for i in range(n-m+1):
        if haystack[i:i+m] == needle:
            return i
    return -1
```

```python
#brute force
#1. needle empty, haystack empty, return 0
def strstr(haystack, needle):
    m = len(needle)
    n = len(haystack)
       
    if m == 0:
        return 0
    if n == 0:
        return -1
    if m > n:
        return -1

    i = j = 0
    start = i
    while i < n and j < m:
        if needle[j] == haystack[i]:
            #print(i, j)
            i += 1
            j += 1
            if (j == m ) and (i < n+1):
                return start
        else:
            i += 1
            j = 0
            start = i
    return -1
            
```

# Leetcode 720. Longest Word in Dictionary (Medium)

Given an array of strings words representing an English Dictionary, return the longest word in words that can be built one character at a time by other words in words.

If there is more than one possible answer, return the longest word with the smallest lexicographical order. If there is no answer, return the empty string.

Note that the word should be built from left to right with each additional character being added to the end of a previous word. 

- Example 1:

Input: words = ["w","wo","wor","worl","world"], Output: "world"

Explanation: The word "world" can be built one character at a time by "w", "wo", "wor", and "worl".

- Example 2:

Input: words = ["a","banana","app","appl","ap","apply","apple"],Output: "apple"

Explanation: Both "apply" and "apple" can be built from other words in the dictionary. However, "apple" is lexicographically smaller than "apply".

```python
def longestWord(slist):
    myDict = {}
    slist.sort()
    result = ""
    n = len(slist)
    for s in slist:
        myDict[s] = len(s)
    for current in slist[::-1]:
        flag = 0
        m = len(current)
        for j in range(1,m):
            if current[0:m-j] not in myDict:                
                flag = 1
                break
        if flag == 0 and len(current) > len(result):
            result = current
        elif flag == 0 and len(current) == len(result):
            result = min(result, current)
        elif flag == 0:
            break
    return result
```

```python
longestWord(["a","banana","app","appl","ap","apply","apple"])
```

```python
longestWord(["w","wo","wor","worl","world"])
```

# Leetcode 151. Reverse Words in a String (Medium)

Given an input string s, reverse the order of the words.

A word is defined as a sequence of non-space characters. The words in s will be separated by at least one space.

Return a string of the words in reverse order concatenated by a single space.

Note that s may contain leading or trailing spaces or multiple spaces between two words. The returned string should only have a single space separating the words. Do not include any extra spaces.

- Example 1:

Input: s = "the sky is blue"
Output: "blue is sky the"

- Example 2:

Input: s = "  hello world  "
Output: "world hello"
Explanation: Your reversed string should not contain leading or trailing spaces.

- Example 3:

Input: s = "a good   example"
Output: "example good a"
Explanation: You need to reduce multiple spaces between two words to a single space in the reversed string.

```python
def reverse_words(myStrings):
    if not myStrings:
        return []
    reversed_list =''
    n = len(myStrings)
    j = n
    for i in range(n-1, -1, -1):
        if myStrings[i] == ' ':
            j = i
        elif (i == 0 or myStrings[i-1] == ' '):
            if len(reversed_list) != 0:
                reversed_list +=' '
            reversed_list += (myStrings[i:j])
    return reversed_list   
```

```python
reverse_words(" hello world")
```

# Leetcode 186. Reverse Words in a String II (Medium)

Given a character array s, reverse the order of the words.

A word is defined as a sequence of non-space characters. The words in s will be separated by a single space.

Your code must solve the problem in-place, i.e. without allocating extra space.

- Example 1:

Input: s = ["t","h","e"," ","s","k","y"," ","i","s"," ","b","l","u","e"]
Output: ["b","l","u","e"," ","i","s"," ","s","k","y"," ","t","h","e"]

- Example 2:

Input: s = ["a"]
Output: ["a"]

```python
def reverse2(slist, start, end):
    #reverse the list
    i = start
    j = end
    while i < j:
        slist[i], slist[j] = slist[j], slist[i]
        i += 1
        j -= 1
    return slist

def reverse_word2(slist):
    reverse2(slist, 0, len(slist)-1)
    j = 0
    for i in range(len(slist)):
        if slist[i] == ' ':
            reverse2(slist, j, i-1)
            j = i + 1
    reverse2(slist, j, i)
    return slist
```

```python
reverse_word2(["t","h","e"," ","s","k","y"," ","i","s"," ","b","l","u","e"])
```

# Leetcode 163. Missing Ranges (Medium)

Description
You are given an inclusive range [lower, upper] and a sorted unique integer array nums, where all elements are within the inclusive range.

A number x is considered missing if x is in the range [lower, upper] and x is not in nums.

Return the shortest sorted list of ranges that exactly covers all the missing numbers. That is, no element of nums is included in any of the ranges, and each missing number is covered by one of the ranges.

- Example 1:

Input: nums = [0,1,3,50,75], lower = 0, upper = 99
Output: [[2,2],[4,49],[51,74],[76,99]]

Explanation: The ranges are:
[2,2]
[4,49]
[51,74]
[76,99]

- Example 2:

Input: nums = [-1], lower = -1, upper = -1
Output: []

Explanation: There are no missing ranges since there are no missing numbers.

```python
def missingRange(nlist, lower, upper):
    i = lower
    j = upper
    n = len(nlist)
    result = []
    if n == 0:
        return [[lower, upper]]
    if lower < nlist[0]:
        result.append([lower, nlist[0] -1])
    k = 1
    while k <  n :
        if (nlist[k-1] + 1 ) < nlist[k]:
            result.append([nlist[k-1] + 1, nlist[k] - 1])
        k = k + 1
    
    if upper > nlist[n-1]:
        result.append([nlist[n-1]+1, upper])
    return result
```

```python
missingRange([0,1,3,50,75], 0, 99)
```

```python
missingRange([-1], -1, -1)
```

```python
missingRange([], 1, 2)
```

```python
missingRange([1,2], 1, 2)
```

# Leetcode 159. Longest Substring with At Most Two Distinct Characters (Medium)

Description
Given a string s, return the length of the longest substring that contains at most two distinct characters.

- Example 1:

Input: s = "eceba"
Output: 3
Explanation: The substring is "ece" which its length is 3.

- Example 2:

Input: s = "ccaabbb"
Output: 5
Explanation: The substring is "aabbb" which its length is 5.

```python
def longestTwoDistinct(myString):
    if not myString:
        return 0
    maxLength = 0

    for i in range(len(myString)):
        current = i
        charList = []
        #print('i:', i)
        while current < len(myString):  
            #print('current:', current)
            #print('charList:', charList)
            if myString[current] not in charList and len(charList)< 2:
                charList.append(myString[current])
                current += 1
            elif myString[current] not in charList and len(charList) >= 2:
                break
            else:
                current += 1
        length = current - i
        if length > maxLength:
            maxLength = length
    return maxLength    
            
```

```python
longestTwoDistinct("ccaabbb")
```

```python
longestTwoDistinct("eceba")
```
