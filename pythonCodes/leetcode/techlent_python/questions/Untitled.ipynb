# leetcode 136. Single Number

Given a non-empty array of integers, every element appears twice except for one. Find that single one.

Note:
Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

- Example: Input: [2,2,1], Output: 1


```python
# use math
def singleNumber(nlist):
    mySet = set(nlist)
    return sum(mySet)*2 - sum(nlist)
```

```python
# use dictionary
def singleNumber2(nlist):
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict.keys():
        if myDict[i] == 1:
            return i
```

```python
# use
def singleNumber3(nlist):
    if not nlist:
        return None
    result = nlist[0]
    for i in nlist[1:]:
        result = result ^ i
    return result
```

```python
singleNumber3([2,2,1])
```

# leetcode 169. Majority Element

Description:
Given an array of size n, find the majority element. The majority element is the element that appears more than ⌊ n/2 ⌋ times.

Note:
You may assume that the array is non-empty and the majority element always exist in the array.

- Example:
  Input: [3,2,3], Output: 3

```python
def majorityElement(nlist):
    if not nlist:
        return None
    n = len(nlist)
    myDict = {}
    for i in nlist:
        myDict[i] = myDict.get(i, 0) + 1
    for i in myDict.keys():
        if myDict[i] > n//2:
            return i
```

```python
majorityElement([3,2,3])
```

# leetcode 217. Contains Duplicate

- Description:
Given an array of integers, find if the array contains any duplicates.

Note:
Your function should return true if any value appears at least twice in the array, and it should return false
if every element is distinct.

- Example: Input: [1,2,3,1], Output: true

```python
def containDuplicate(nlist):
  if not nlist:
    return False
  myDict = {}
  for i in nlist:
    if i in myDict:
      return True
    else:
      myDict[i] = myDict.get(i, 0)+1
  return False
```

```python
containDuplicate([1,2,3,1])
```

```python
containDuplicate([1,2,3,4])
```
