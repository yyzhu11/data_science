# Method1: Use Functions
### apply( ), map( ) ★★★★★

```python
def breakup(x):
    if type(x) == str: # only works for string input
        parts = x.split(',')
        return parts
```

#### For product_type

```python
dict_type = {}
for product in df[df.product_type.notnull()].product_type.map(breakup):
    for p_type in product:
        if p_type in dict_type.keys(): 
            dict_type[p_type] += 1
        else:
            dict_type[p_type] = 1
```

```python
len(dict_type)
```

```python
sorted(dict_type.items(), key = lambda x: x[1], reverse = True)
```

### For maker

```python
dict_maker = {}
for maker in df[df.maker.notnull()].maker.map(breakup):
    for m in maker:
        if m in dict_maker.keys():
            dict_maker[m] += 1
        else:
            dict_maker[m] = 1
```

```python
len(dict_maker)
```

```python
sorted(dict_maker.items(), key = lambda x: x[1], reverse = True)
```

### For ingredient

```python
dict_ingredient = {}
for ingred in df[df.ingredient.notnull()].ingredient.map(breakup):
    for i in ingred:
        if i in dict_ingredient.keys():
            dict_ingredient[i] += 1
        else:
            dict_ingredient[i] = 1
```

```python
len(dict_ingredient)
```

```python
sorted(dict_ingredient.items(), key = lambda x: x[1], reverse = True)
```

# Method 3: Use Explode built-in function

```python
df_test = pd.DataFrame({'product_type': [['O','B'],['B']], 'dummy': 1})
print("------The DataFrame is--------")
print(df_test)
print("After expanding the DataFrame")
print(df_test.explode('product_type'))
print('Value counts')
print(df_test.explode('product_type').product_type.value_counts())
```
